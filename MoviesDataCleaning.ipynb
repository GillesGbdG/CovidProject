```python
# import libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of the plots we will create

#Read in the data

df = pd.read_csv(r'C:\Users\gille\OneDrive\Documents\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# let's look for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} -{}%'.format(col,pct_missing))
   # print(df[col].isnull().value_counts(), "\n")
```

```python
# columns Data types

df.dtypes 
```

```python
# Change data type of budget, gross and votes columns

df = df.dropna() # delete all the rows with missing values 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64') 
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
# create new year column 

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Let's look at the data
df
```

```python
# Sort by gross values 

df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for movie')
plt.show()
```

```python
# plot the bufget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "green"})
```

```python
#let's look at the correlation 
df.corr(method= 'pearson') #pearson, kendall, spearman
```

```python
# High correlation bettwen budget and gross

correlation_matrix = df.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlatiom Matrix for numeric features')
plt.xlabel('Movies info')
plt.ylabel('Movies info')

plt.show()
```

```python
# convert string values to category type

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
    
    
```

```python
correlation_matrix = df_numerized.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlatiom Matrix for numeric features')
plt.xlabel('Movies info')
plt.ylabel('Movies info')

plt.show()
```
